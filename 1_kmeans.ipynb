### 1. Descripción

K-means es un algoritmo de clustering (agrupamiento) particional, lo que significa que divide un conjunto de datos en grupos o clústeres basados en sus similitudes, donde K es un número predefinido por el usuario.
El objetivo principal es minimizar la suma de las distancias al cuadrado entre los puntos de datos y los centroides de los clústeres a los que pertenecen.

### 3. Tipo de Modelo:

* **Método de Aprendizaje:** Algoritmos de Machine Learning NO Supervisados ya que no utiliza ejemplos de entrenamiento etiquetados para ajustar un modelo, si no que se basa en el proceso de asignación iterativa y actualización de centroides para formar los clústeres.

*  **Por Parámetro:** . El algoritmo K-means clustering no es parametrizado en el sentido de que no se basa en un modelo con parámetros ajustables que se optimizan durante el proceso de entrenamiento, como lo hacen los algoritmos supervisados.

* **Dato de Prendizaje:**  OFF-LINE donde opera en un enfoque que no considera datos en tiempo real de manera continua, K-means trabaja con un conjunto de datos fijo y realiza su proceso de asignación y actualización de centroides en una sola ejecución, sin incorporar nuevos datos de manera incremental. El proceso de asignación y actualización de centroides se basa en la distancia entre los puntos de datos.

* **Resultado del Entrenamiento:** El resultado del entrenamiento no es un Modelo, K-means no crea un modelo que pueda ser utilizado para predecir etiquetas o clasificar nuevos datos. K-means es una asignación de puntos de datos a los clústeres finales y los centroides de esos clústeres, na vez que se completa el proceso de entrenamiento se puede utilizar los clústeres formados para analizar y comprender la estructura de tus datos

### 5. Supuestos y Restricciones:

* **Número de Clústeres Predefinido:** El usuario debe especificar previamente el número de clústeres en los que desea dividir los datos. Esto puede ser problemático si no se conoce la cantidad óptima de clústeres, ya que elegir un número incorrecto puede dar lugar a agrupamientos inadecuados.

* **Sensible a la Inicialización:** El resultado del algoritmo K-Means puede depender de la ubicación inicial de los centroides de los clústeres. Iniciar los centroides de manera subóptima podría llevar a convergencia en mínimos locales en lugar del mínimo global óptimo.

* **Forma y Tamaño del Clúster:** K-Means supone que los clústeres son esféricos y de tamaño similar. Esto significa que puede no funcionar bien con datos donde los clústeres tienen formas irregulares o tamaños muy diferentes.

* **Sensibilidad a los Valores Atípicos:** Los valores atípicos (outliers) pueden afectar significativamente el resultado del algoritmo, ya que pueden influir en la ubicación de los centroides y, por lo tanto, en los límites de los clústeres.

* **Distancias Euclidianas:** K-Means utiliza distancias euclidianas para medir la similitud entre puntos y centroides. Esto puede ser problemático en datos de alta dimensionalidad, ya que la noción de distancia euclidiana puede volverse menos significativa en espacios de alta dimensión (fenómeno conocido como "maldición de la dimensionalidad").

* **Número de Puntos por Clúster Desigual:** K-Means asume que los clústeres tienen una distribución uniforme de puntos, lo que significa que todos los clústeres deben tener aproximadamente el mismo número de puntos. Esto puede no ser adecuado para datos donde los clústeres tienen tamaños muy diferentes.

* **No Garantiza la Convergencia Global:** El algoritmo K-Means puede converger a un mínimo local en lugar del mínimo global, dependiendo de la inicialización de los centroides.

* **Dependencia de Escala:** El algoritmo es sensible a las escalas de las características. Características con diferentes escalas pueden dominar la distancia euclidiana, lo que lleva a resultados sesgados.

* **No Maneja Bien los Datos Categóricos:** K-Means se basa en medidas de distancia, por lo que no es adecuado para datos categóricos, a menos que se utilicen transformaciones adecuadas.

### 6. Ejemplo

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Generate some sample data
n_samples = 300
n_features = 2
n_clusters = 3

X, _ = make_blobs(n_samples=n_samples, n_features=n_features, centers=n_clusters, random_state=42)

# Create and fit the K-Means model
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
kmeans.fit(X)

# Get cluster assignments and cluster centers
cluster_labels = kmeans.labels_
cluster_centers = kmeans.cluster_centers_

# Plot the data points and cluster centers
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=cluster_labels, cmap='viridis', s=50)
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='red', marker='x', s=200, label='Cluster Centers')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-Means Clustering')
plt.legend()
plt.show()

# Print cluster assignments
print("Cluster assignments:")
print(cluster_labels)
```
