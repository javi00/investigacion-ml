### 1. Descripción

El algoritmo de agrupamiento de k-mediana es una variación del algoritmo de k-medias que utiliza la mediana de los puntos de datos en cada grupo como centro del grupo en lugar de la media (promedio) utilizada en k-medias. Esto puede hacerlo más robusto ante los valores atípicos en los datos.

### 3. Tipo de Modelo

* **Método de Aprendizaje** : K-median es un algoritmo de aprendizaje **no supervisado**. No se basa en datos etiquetados ni en valores objetivo para el entrenamiento; su objetivo es descubrir patrones y estructuras en datos no etiquetados.
* **Dato de Prendizaje**: K-median suele considerarse un algoritmo de aprendizaje **por lotes**. Procesa todo el conjunto de datos a la vez y calcula las medianas de los conglomerados basándose en todos los puntos de datos. No se adapta a los nuevos puntos de datos de forma incremental (en línea) después de la agrupación inicial.
* **Basado en instancias o en modelos**: K-median es un algoritmo **basado en instancias**. No crea un modelo en el sentido tradicional, como la regresión lineal o las redes neuronales. En su lugar, asigna directamente cada punto de datos a un conglomerado en función de su proximidad a las medianas de los conglomerados. No construye un modelo global que pueda hacer predicciones.
* **Paramétrico o no paramétrico**: K-median se considera un algoritmo **no paramétrico**. No hace suposiciones específicas sobre la distribución de datos subyacente. 

### 4. Algoritmos de entrenamiento

El algoritmo de agrupamiento de mediana k es un algoritmo tiene como objetivo dividir un conjunto de datos en grupos de una manera que minimice la distancia media entre los puntos de datos y el centro del grupo (mediana). Estos son los pasos clave y los detalles del algoritmo de entrenamiento:

* Inicialización: el algoritmo comienza seleccionando aleatoriamente k centros de conglomerados iniciales (medianas) del conjunto de datos, donde k es un parámetro definido por el usuario que representa el número deseado de conglomerados.

* Paso de la tarea: para cada punto de datos en el conjunto de datos, calcule la distancia entre el punto de datos y los k centros de conglomerados. Asigne el punto de datos al grupo con la mediana más cercana (normalmente utilizando una métrica de distancia como la distancia euclidiana). Este paso da como resultado la formación de k grupos iniciales.

* Paso de actualización: vuelva a calcular las medianas (centros) de los grupos en función de los puntos de datos asignados a cada grupo. Las nuevas medianas de los grupos se establecen como las medianas de todos los puntos de datos dentro de los respectivos grupos.

* Comprobación de convergencia: compruebe si el algoritmo ha convergido. La convergencia se puede determinar en función de varios criterios, como cuando las asignaciones de los conglomerados ya no cambian o cuando el cambio en las medianas de los conglomerados cae por debajo de un cierto umbral.

* Repita los pasos 2 a 4: si no se ha alcanzado la convergencia, repita los pasos de asignación y actualización hasta que el algoritmo converja. Vale la pena señalar que la k-mediana no garantiza un óptimo global y, en ocasiones, el algoritmo puede converger a un mínimo local.

* Salida: El resultado final del algoritmo de k-mediana es un conjunto de k medianas de grupos y la asignación de puntos de datos a estos grupos. Cada punto de datos pertenece al grupo cuya mediana está más cercana a él.

* Parámetros: los parámetros clave que deben establecerse antes de aplicar el algoritmo de mediana k incluyen el número de grupos (k) y la elección de la métrica de distancia para medir la similitud entre los puntos de datos (por ejemplo, la distancia euclidiana).

* Complejidad: el algoritmo de k-mediana puede ser computacionalmente intensivo, especialmente para conjuntos de datos grandes, ya que implica calcular distancias entre puntos de datos y medianas de grupos en cada iteración. Existen varias variaciones y técnicas de optimización para mejorar su eficiencia.

Es importante tener en cuenta que la k-mediana es un algoritmo de agrupamiento heurístico y la calidad de la solución de agrupamiento obtenida puede depender de la selección aleatoria inicial de los centros del grupo y de la métrica de distancia específica utilizada. Se pueden explorar diferentes variaciones de k-mediana y estrategias de inicialización para encontrar los mejores resultados de agrupación para un conjunto de datos determinado.

### 5. Supuestos y restricciones

Supuestos:

* Métrica de distancia euclidiana: la mediana K normalmente supone que la distancia entre puntos de datos se mide utilizando la métrica de distancia euclidiana. Si bien se puede adaptar a otras métricas de distancia, esta es una suposición común.

* Forma de los conglomerados: La mediana K supone que los conglomerados tienen forma esférica o aproximadamente esférica. Esto significa que es posible que no funcione bien en conjuntos de datos con grupos de formas irregulares o grupos alargados.

* Número de conglomerados (k): el algoritmo supone que el número de conglomerados (k) es conocido o está preespecificado. Seleccionar un valor apropiado para k puede ser un desafío y elegir un valor incorrecto puede generar resultados subóptimos.

* Tamaños de conglomerados homogéneos: la mediana K supone que los conglomerados tienen aproximadamente la misma cantidad de puntos de datos o que la variación del tamaño del conglomerado no es una preocupación significativa. Si los tamaños de los clústeres están muy desequilibrados, es posible que el algoritmo no funcione bien.

* Inicialización: el rendimiento de la mediana K puede ser sensible a la selección inicial de los centros del clúster. El algoritmo supone que se utiliza una estrategia de inicialización adecuada para elegir las medianas iniciales.

Restricciones:

* Óptimo local: la mediana K es un algoritmo heurístico y puede converger a óptimos locales en lugar de al óptimo global. Diferentes ejecuciones del algoritmo con diferentes inicializaciones pueden dar como resultado diferentes agrupaciones.

* Escalabilidad: la complejidad computacional de la k-mediana aumenta con el tamaño del conjunto de datos. Puede resultar computacionalmente costoso para conjuntos de datos muy grandes o datos de alta dimensión.

* Valores atípicos: la mediana K puede ser sensible a los valores atípicos en el conjunto de datos. Los valores atípicos pueden afectar desproporcionadamente la ubicación de los centros de los clusters y los resultados generales del clustering.

* Clústeres no convexos: la mediana K está diseñada para conglomerados convexos. Si el conjunto de datos contiene grupos no convexos, la k-mediana puede tener dificultades para dividir los datos con precisión.

* Distribución de datos: K-median no hace suposiciones sobre la distribución de datos subyacente. Si bien esto puede ser una ventaja en muchos casos, también puede generar resultados subóptimos si la distribución de los datos no se alinea con los supuestos de la mediana k.

* Determinista: K-mediana es un algoritmo determinista, lo que significa que producirá los mismos resultados dadas las mismas condiciones iniciales. Esto puede ser una limitación cuando se trata de datos ruidosos o inciertos.

### 6. Ejemplo

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from pyclustering.cluster.kmedians import kmedians
from pyclustering.cluster import cluster_visualizer

# Generate a synthetic dataset with three clusters
n_samples = 300
n_features = 2
n_clusters = 2
random_state = 42

data, labels = make_blobs(n_samples=n_samples, n_features=n_features, centers=n_clusters, random_state=random_state)

# Convert the data to PyClustering-compatible format
data_points = np.array(data)

# Generate initial medians by randomly selecting data points
initial_medians = data_points[np.random.choice(range(len(data)), n_clusters)]

# Create the kmedians instance with initial medians
kmedians_instance = kmedians(data_points, initial_medians)

# Run the clustering algorithm
kmedians_instance.process()
clusters = kmedians_instance.get_clusters()
medians = kmedians_instance.get_medians()

# Create a single figure
plt.figure(figsize=(8, 6))

# Visualize the clusters
for cluster_id, cluster in enumerate(clusters):
    cluster_points = data_points[cluster]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], label=f'Cluster {cluster_id + 1}')

# Manually add labels for medians
for i, median in enumerate(medians):
    plt.scatter(median[0], median[1], c='red', marker='x', s=200)
    plt.annotate(f'Median {i + 1}', (median[0], median[1]), textcoords="offset points", xytext=(0, 10), ha='center')

# Set labels and legend
plt.title('K-Medians Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()

# Show the plot
plt.show()

# Print cluster medians
print("Cluster Medians:")
for i, median in enumerate(medians):
    print(f"Median {i + 1}: {median}")
```
